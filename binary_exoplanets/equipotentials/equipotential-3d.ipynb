```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
import equipotentials
```

```python
eq = equipotentials.Equipotentials(0.1, 400)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(eq.X, eq.Y, eq.V, cmap="coolwarm",
                       rstride=5, cstride=5, linewidth=0, vmin=1.5, vmax=2.0)
ax.set_zlim(1.0,2.0)
```
